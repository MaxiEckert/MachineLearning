```python
import tensorflow as tf
import numpy as np

import time
import datetime as dt

from tensorflow.examples.tutorials.mnist import input_data

tf.reset_default_graph()

# input vector of type scalar with length 2
#x = tf.placeholder(tf.float32, [None, 2], name='x')  #simple 2D data
x = tf.placeholder(tf.float32, [None, 784], name='x')   #MNIST
# target labels
#target = tf.placeholder(tf.float32, [None, 1], name='target')    #simple 0-1 classification
target = tf.placeholder(tf.float32, [None, 10], name='target')     #MNIST 10 classes

# bias
#b = tf.get_variable("b", [1])   #2D data
b = tf.get_variable("b", [10])    #MNIST

# weights
#w = tf.get_variable("w", [1,2]) # weight matrix W for simple 2D
w = tf.get_variable("w", [784,10]) # weight matrix W for MNIST

dot_product = tf.add(tf.matmul(x, w), b)

output = tf.sigmoid(dot_product)
#output = tf.sign(dot_product)

mse = tf.reduce_mean(tf.square(target - output))

learning_rate = 0.4

# Define a training step using Gradient Decent to minimize the mean squared error
optimizer = tf.train.AdagradOptimizer(learning_rate=learning_rate).minimize(mse)
#optimizer = tf.train.GradientDescentOptimizer(learning_rate=learning_rate).minimize(mse)

# training data
input = np.array([[1,1],[1,2],[2,1],[1,4],[2,3],[3,3],[4,2]])
labels = np.array([[0],[0],[0],[1],[1],[1],[1]])

# MNIST training data
mnist = input_data.read_data_sets("MNIST_data/", one_hot=True)


# ------------------------
# Visualization of flow graph
# Define summaries to be shown in TensorBoard
tf.summary.histogram("Weights", w)
tf.summary.scalar("Bias", b[0])
tf.summary.scalar("MSE", mse)

# ------------------------------
# Activate logging for TensorBoard
merged = tf.summary.merge_all()
st = dt.datetime.fromtimestamp(time.time()).strftime('%Y-%m-%d_%H-%M-%S')
train_writer = tf.summary.FileWriter('log/lr_' + st, tf.get_default_graph())
# ------------------------------


iterations = 10000
init = tf.global_variables_initializer()

with tf.Session() as sess:
    init.run()
    
    #for j in range(iterations):
    # Loop over all steps and perform training updates on W and b
    for i in range(iterations):      
        # Instead of MiniBatches we can use the whole set here
        xs = np.array(input)   # input
        ys = np.array(labels)  # blood pressure
        
        #MNIST
        batch_xs, batch_ys = mnist.train.next_batch(100)
        
        #summary, _ = sess.run([merged, optimizer], feed_dict={ x: xs, target: ys })
        res_mse, res_w, res_b, summary, _ = sess.run([mse, w, b, merged, optimizer], feed_dict={ x: batch_xs, target: batch_ys })
        train_writer.add_summary(summary, i)

        #res_mse, res_w, res_b = sess.run([mse, w, b], feed_dict={ x: xs, target: ys})
        print("\nAfter iteration: %d" % i)
        #print("MSE: %f \n" % res_mse)
        print(res_mse)
        #print(res_b)
        #print(w.eval())
        
        
    # -----------------------
    # Test/Evaluation
    correct_prediction = tf.equal(tf.argmax(output,1), tf.argmax(target,1))
    accuracy = tf.reduce_mean(tf.cast(correct_prediction, tf.float32))
    print("\n\n-------------------------------\n")
    print("Overall accuracy: ")
    print(sess.run(accuracy, feed_dict={x: mnist.test.images, target: mnist.test.labels}))

        
```
